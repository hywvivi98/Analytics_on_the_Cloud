```scala
//PROBLEM 1 SOLUTION
val test_data = "widgets 103.24,eggs 345.22,milk 231.25,widgets 123.11,milk 14.2" 

//YOUR FUNCTION SHOULD GO HERE
def inventory_value(test_data:String) = 
    test_data
              .toLowerCase.split(",") //split with comma
              .map(d=> (d.split(' '))).map(v =>( v(0),v(1))).sortWith(_._1 < _._1).groupBy(_._1) //group by product names
              .mapValues(i=>i.map(j=>j._2.toDouble).reduce(_+_)) // find sum of each category



//Test your function here
inventory_value(test_data)
```

```scala
//PROBLEM 2 SOLUTION
val test_data = "able was i ere i saw elba, and that was before i was able and could see elba!"

//YOUR FUNCTION SHOULD GO HERE
def letter_count(test_data:String) = 
    test_data
              .replaceAll("[^A-Za-z]+", "").toLowerCase.toArray //convert to char array
              .map(w=>(w,1)).groupBy(_._1) //map and group by category
              .map(w => (w._1,w._2.map(_=>1).reduce(_+_))) // sum up


//Test your function here
letter_count(test_data)
```

```scala
//PROBLEM 3 SOLUTION
val sample_essay = """able was i ere i saw elba
a man a plan a canal panama
the future is uncertain and the end is always near"""

//Your function should go here
def essay_grade(sample_essay:String) : Int = {
      val pct = sample_essay
                    .linesIterator.map(_.trim).toArray // convert sample essay to lines of string
                    .map(line => line.count((( c: Char ) => List( 'a', 'e', 'i','o','u').contains( c ))).toDouble / line.size) // get percent of vowel in the string
     ((1 - pct.reduce(_ + _) / pct.size) * 100).toInt } // return final score

//Test your function here
essay_grade(sample_essay)
```
